```python
import numpy as np
import pandas as pd
```

```python
path = './dataset/'

plants_file = 'plants.csv'
plants_clean_file = 'plants_clean.csv'
```

```python
plants = pd.read_csv(path + plants_file, encoding='cp1252', skip_blank_lines=True)
plants_clean = pd.read_csv(path + plants_clean_file, encoding='cp1252', skip_blank_lines=True)
```

```python
print(plants['Plant Name'].duplicated().sum(), "doublons trouvés :")
```

```python
plants_clean = plants.drop_duplicates(subset='Plant Name', keep='first')
```

```python
print(plants_clean['Plant Name'].duplicated().sum(), "doublons trouvés après nettoyage")
```

```python
plants_clean.to_csv(path + plants_clean_file, index=False)
```
